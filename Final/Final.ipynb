# Final changes in the dataset.

### Importing necessary libraries and the dataset.

```python
import pandas as pd
import numpy as np
```

```python
df_titanic = pd.read_csv("NewTitanic.tsv", sep='\t')
```

### I am changing the datatypes in the Age column to categorize them. Because in Orange I want to use Age info more efficiently.

```python
df_titanic['Age'].unique()
```

### Here, I am changing the text values to numerical ones.

```python
df_titanic['Age'] = df_titanic['Age'].replace(['NoInfo', 'WrongValue'], ['0.0', '-1.0'])
```

```python
df_titanic['Age'].unique()
```

### Then changing the datatype.

```python
df_titanic['Age'] = df_titanic['Age'].astype(float)
```

```python
df_titanic['Age'].unique()
```

```python
df_titanic.head(5)
```

```python
df_titanic.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_titanic.head(5)
```

### Changing the PClass Column to int because I want to see them with numbers.

```python
df_titanic['Pclass'] = df_titanic['Pclass'].replace(['First', 'Second', 'Third'], ['1', '2', '3'])
```

```python
df_titanic['Pclass'] = df_titanic['Pclass'].astype(int)
```

```python
df_titanic['Survived'] = df_titanic['Survived'].astype(int)
```

```python
df_titanic['Age'].describe()
```

### Here, I am categorizing the Age column because I want to find Young examples to answer the questions. So it will be easy to see Ages categorized.

```python
category = ['NoInfo', 'Child', 'Young', 'Adult', 'Senior']

df_titanic['Ages'] = pd.cut(x=df_titanic['Age'], bins=[-2,0,14,24,64,100], labels=category)
```

```python
df_titanic.rename(columns={"Ages": "AgeCategory"})
```

### And finally I am changing SibSp column values to 0 or 1 instead of numbers that they have because it is easy to understand this way if they have a family with them or not.

```python
df_titanic['SibSp'] = df_titanic['SibSp'].astype(bool).astype(int)
```

```python
df_titanic['Sex'].unique()
```

### Then saving the dataset to a new one.

```python
df_titanic.to_csv('FinalTitanic.tsv', sep='\t')
```
